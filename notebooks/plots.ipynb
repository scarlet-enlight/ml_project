```python
import pandas as pd

df = pd.read_csv(r"../data/processed/Sleep_health_and_lifestyle_dataset.csv")
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plot = sns.heatmap(df.corr(), annot=True)
plt.show(plot)
```

```python
plt.figure(figsize=(8,3))
sns.histplot(df['Sleep Disorder'], bins=3, kde=True)
plt.xlabel('Sleep Disorder (0 - None, 1 - Sleep Apnea, 2 - Insomnia)')
plt.ylabel('Amount among testers')
plt.title('Number of people having sleep disorders') 
plt.show()
```

```python
plt.figure(figsize=(8,3))
sns.boxplot(y='Sleep Duration', x='Sleep Disorder', data=df, palette='Set2')
plt.title("Sleeping Disorders proportion to testers' average times of sleeping")
plt.xlabel('Sleep Disorder (0 - None, 1 - Sleep Apnea, 2 - Insomnia)')
plt.ylabel('Sleep Duration (in hours)')
plt.show()
```

```python
sns.lineplot(x="Age", y="Sleep Duration", data=df)
```

```python
g = sns.catplot(
    data=df, kind="bar",
    x="Gender", y="BMI Category", hue="Stress Level"
)
g.despine(left=True)
g.set_axis_labels("Male to Female", "BMI Category")
g.legend.set_title("Ratio of Male and Female to BMI Category with inclusion of stress levels (3 to 8)")
```
